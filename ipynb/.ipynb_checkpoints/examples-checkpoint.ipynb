```python
# Get a recent Auckland GTFSR trip update feed
api_key = os.environ['AT_KEY']
url = 'https://api.at.govt.nz/v2/public/realtime/?'#tripupdates?'
headers = {
  'Ocp-Apim-Subscription-Key': api_key,
  }
params = {
  'callback': '',
  'tripid': '',
  'vehicleid': '',
  }    

r = requests.get(url, headers=headers, params=params)
feed = r.json()['response']
```

```python

# Fix Auckland feeds on file



def fix_times(s):
    t = {}
    if isinstance(s, dict):      
        for k, v in s.items():
            if isinstance(v, dict):
                if k == 'stop_time_update':
                    t[k] = [fix_times(v)]
                else:
                    t[k] = fix_times(v)
            elif isinstance(v, list):
                t[k] = [fix_times(x) for x in v]
            elif k in ['timestamp', 'time']:
                t[k] = int(v)
            else:
                t[k] = v 
    elif isinstance(s, list):
        t = [fix_times(x) for x in s]
    else:
        t = s
    return t


path = DATA_DIR/'test_gtfsr'/'20160519080020.json'
with path.open() as src:
    feed = json.load(src)
    
feed = fix_times(feed)
path = DATA_DIR/'test_gtfsr'/'20160519080020_t.json'
with path.open('w') as tgt:
    json.dump(feed, tgt)

```
